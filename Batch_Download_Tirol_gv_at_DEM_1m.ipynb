```python
import urllib2, datetime, dateutil.rrule, os
import pandas as pd
import glob
import zipfile
import shutil
```

```python
listOfUrls=pd.read_csv('../DGM_DOM_list_CH.csv',sep=',')
```

```python
listOfUrls.head()
```

```python
for url in listOfUrls['URL']:
    try:
        response=urllib2.urlopen(url)
        data=response.read()
        with open('../data/%s'%(url.split('/')[-1]),'w') as outfile:
            outfile.write(data)
        print('wrote %s'%(url.split('/')[-1]))
    except:
        print('error')
```
